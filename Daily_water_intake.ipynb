```python
data['Gender']=data['Gender'].replace({'Male':0, 'Female':1})
```

```python
data['Physical Activity Level'].unique()
```

```python
data['Physical Activity Level']=data['Physical Activity Level'].replace({'Moderate':0, 'High':1, 'Low':2})
```

```python
data['Weather'].unique()
```

```python
data['Weather']=data['Weather'].replace({'Hot':0, 'Normal':1, 'Cold':2})
```

```python
data['Weather'].unique()
```

```python
data['Hydration Level'].unique()
```

```python
data['Hydration Level']=data['Hydration Level'].replace({'Good':1, 'Poor':0})
```

```python
data['Hydration Level'].unique()
```

```python
data.info()
```

```python
features=data.drop('Hydration Level',axis=1)
```

```python
target=data['Hydration Level']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,target,random_state=42,train_size=0.75)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_predict=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```
